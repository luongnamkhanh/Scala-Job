```python
from datetime import datetime, timedelta
import random

def generate_random_datetime_string(start_date, end_date, date_format='%Y-%m-%d %H:%M:%S'):
    # Convert start and end date strings to datetime objects
    start_datetime = datetime.strptime(start_date, date_format)
    end_datetime = datetime.strptime(end_date, date_format)
    
    # Calculate a random datetime within the specified range
    random_datetime = start_datetime + timedelta(seconds=random.randint(0, int((end_datetime - start_datetime).total_seconds())))
    
    # Format the random datetime as a string
    random_datetime_string = random_datetime.strftime(date_format)
    
    return random_datetime_string

# Example usage:
start_date = '2023-01-01 00:00:00'
end_date = '2023-12-31 23:59:59'
date_time = []
for _ in range(100000):
    date_time.append(generate_random_datetime_string(start_date, end_date))

print(date_time)
    
```

```python
from datetime import datetime

def get_day_of_week(date_string):
    try:
        # Convert the date string to a datetime object
        date_obj = datetime.strptime(date_string, '%Y-%m-%d %H:%M:%S')

        # List of day names
        day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

        # Get the day of the week as an integer (0 = Monday, 1 = Tuesday, ..., 6 = Sunday)
        day_of_week_int = date_obj.weekday()

        # Get the day name by indexing the list using the day_of_week_int
        day_of_week_name = day_names[day_of_week_int]

        return day_of_week_name
    except ValueError:
        return "Invalid Date Format"

# Example usage:
date_string = '2023-10-01 14:25:01'
day_of_week = get_day_of_week(date_string)
print("Day of the Week:", day_of_week)
```

```python
import socket
import struct
import random

def generate_ip():
    # Generate a random IP address in the format "x.x.x.x"
    ip = socket.inet_ntoa(struct.pack('>I', random.randint(1, 0xffffffff)))
    return ip

ip = []
for _ in range(100000):
    ip.append(generate_ip())
```

```python
import requests as rq
from bs4 import BeautifulSoup
# Create an empty dictionary
url_dict = {'cafef.vn': {}, 'genk.vn':{}, 'autopro.com.vn': {}, 'kenh14.vn':{},'afamily.vn':{},'soha.vn':{},'cafebiz.vn':{}}

url_dict_config = {
    'cafef.vn': ['.sub-cat a','h3'],
    'genk.vn': ['.gbh-menu-list a','h4'],
    'autopro.com.vn':['.bottom a','h3'],
    'kenh14.vn':['.khw-bottom-header a','h3'],
    'afamily.vn': ['.afwmw-header a', 'h3'],
    'soha.vn': ['.page-menu a', 'h3'],
    'cafebiz.vn': ['.menu a', 'h3']
}
# Send a GET request to the URL
for _ in url_dict.keys():
    response = rq.get(f'https://{_}')

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

   
        # Find all the <a> elements within the <div> with class="sub-cat"
        links = soup.select(url_dict_config[_][0])

        # Extract and add the href attributes to the dictionary under the 'cafef' key
        categories = []
        for link in links:
            if str(link['href']).endswith('.chn') or str(link['href']).endswith('.htm'):
                categories.append(link['href'])
        

        for category in categories:
            res = rq.get(f'https://{_}{category}')
            soup = BeautifulSoup(res.text, 'html.parser')
            # Find all header tags
            head_tags = soup.find_all(url_dict_config[_][1])
            # Extract href attributes from <a> tags within header tags and store them in a list
            href_list = [a.get('href') for head in head_tags for a in head.find_all('a')]
            url_dict[_][category] = href_list

    else:
        print('Failed to retrieve the web page. Status code:', response.status_code)

print(url_dict)
```

```python
# Remove categories with empty lists as values
url_dict = {domain: {category: paths for category, paths in categories.items() if paths} for domain, categories in url_dict.items()}
```

```python
import uuid

def generate_guid():
    # Generate a new GUID and return it as a string
    return str(uuid.uuid4())

guid = []
for _ in range(5000):
    guid.append(generate_guid())
```

```python
import ipaddress
import random

def generate_ipv6():
    # Generate a random IPv6 address
    ipv6 = ipaddress.IPv6Address(random.getrandbits(128))
    return str(ipv6)

ipV6 = []
for _ in range(100000):
    ipV6.append(generate_ipv6())
```

```python
def count_paths(dictionary):
    total_paths = 0

    for key, value in dictionary.items():
        if isinstance(value, dict):
            total_paths += count_paths(value)  # Recursively count paths in sub-dictionary
        elif isinstance(value, list):
            total_paths += len(value)  # Count paths in a list

    return total_paths
```

```python
total_paths = count_paths(url_dict)
print("Total Paths:", total_paths)
```

```python
# Create empty lists to store data
domains = []
categories = []
paths = []
referers = []
browser_codes = []
os_codes = []
os_vers =[]
ips = []
loc_ids = []
guids = []
ipV6s = []

# Number of records you want
num_records = 100000  # You can change this to the desired number of records

# Generate random data for each record
for _ in range(num_records):
    domain = random.choice(list(url_dict.keys()))  # Random domain
    category = random.choice(list(url_dict[domain].keys()))  # Random category within the domain
    path = random.choice(url_dict[domain][category])  # Random path within the category
    referer = random.choice(url_dict[domain][category])
    domains.append(domain)
    categories.append(category)
    paths.append(path)
    referers.append(referer)
    browser_codes.append(random.choice(browser_code))
    os_codes.append(random.choice(os_code))
    os_vers.append(random.choice(os_ver))
    ips.append(random.choice(ip))
    loc_ids.append(random.choice(loc_id))
    guids.append(random.choice(guid))
    ipV6s.append(random.choice(ipV6))

# Create a DataFrame from the lists
data = {'dt':date_time,'cookietime': None,'browser_code':browser_codes,'browser_ver':None,'os_code':os_codes,'os_version':os_vers,
        'ip':ips,'loc_id':loc_ids,'domain': domains,'siteId':None,'channelId':None,'path': paths,'referer':referers,'guid':guids,'geo':None,
        'org_ip':None,'pageloadId':None,'screen':None,'d_guid':None,'category': categories,'utm_source':None,
        'utm_campaign':None,'utm_medium':None, 'milis':None, 'ipV6':ipV6s}
pc = pd.DataFrame(data)

# Print the DataFrame
pc
```
